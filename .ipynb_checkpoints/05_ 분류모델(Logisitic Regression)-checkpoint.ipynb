```python
import seaborn as sns
sns.get_dataset_names()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

- survived:    . 1 , 0  .

- sex:    .  0 , 1  .

- embarked:    . <br> S (Southampton), C (Cherbourg), Q (Queenstown) .
- alone:      . 1  , 0    .

```python
#   
prep_data = titanic.drop(columns=['alive', 'who', 'adult_male', 'class', 'embark_town'])
prep_data.drop("deck", axis=1, inplace=True)

prep_data.head()
```

```python
prep_data.info()
```

```python
#   177 , 2 
# 0   survived  891 non-null    int64  
# 3   age       714 non-null    float64
# 7   embarked  889 non-null    object 
```

```python
#  
# embarked()   'na' 
prep_data["embarked"] = prep_data.embarked.fillna('na')

#    
prep_data["age"] = prep_data.age.transform(lambda x: x.fillna(x.mean()))

prep_data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
# LabelEncoder => False 0, True 1
#    => male => 1, female 1

le = LabelEncoder()
prep_data['sex'] = le.fit_transform(prep_data['sex'])
prep_data['alone'] = le.fit_transform(prep_data['alone'])
prep_data.head()
```

```python
import pandas as pd
# embarked   S,Q,C 3   
#   get_dummies  
prep_data = pd.get_dummies(prep_data, prefix='', prefix_sep='').astype(int)
prep_data.head()
```

```python
import pandas as pd

#   
data = {
    'color': ['red', 'blue', 'green', 'blue', 'red'],
    'size': ['S', 'M', 'L', 'M', 'S'],
    'price': [15.5, 20.0, 14.0, 19.0, 17.5]
}

df = pd.DataFrame(data)
print(df)
print("-------------------------------------------------------------")
#   - 
df_encoded = pd.get_dummies(df, prefix='', prefix_sep='')

print(df_encoded)
```

```python
from sklearn.model_selection import train_test_split

# (feature) (label) 
# features = prep_data.iloc[:, 1:]  #         
# labels = prep_data.iloc[:, 0]  #      

X_train, X_test, y_train, y_test = train_test_split(prep_data.iloc[:, 1:], 
                                                   prep_data.iloc[:, 0],
                                                   random_state=42)
```

```python
#       (feature)     
#               
#  :   :  :  :
#  ; -  (Min-Max Scaling):,  (Standardization):Robust  (Robust Scaling):
```

```python
from sklearn.preprocessing import RobustScaler
#  (X_train)  (X_test)     
#      
scaler = RobustScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train_scaled, y_train)
#  score  
#     79.82%   .
clf.score(X_test_scaled, y_test)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#  
titanic = sns.load_dataset('titanic')

#    
plt.figure(figsize=(10, 6))
sns.barplot(x='pclass', y='survived', data=titanic)
plt.title('Survival Rate by Class')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
```
